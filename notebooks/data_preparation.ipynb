```python
date_cols = ["PaperlessDate", "InvoiceDate", "DueDate", "SettledDate"]
num_cols = ["countryCode", "invoiceNumber", "InvoiceAmount", "DaysToSettle", "DaysLate"]
cat_cols = ["customerID", "Disputed", "PaperlessBill"]

for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors="coerce", dayfirst=False)
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")
for c in cat_cols:
    df[c] = df[c].astype(str).str.strip()
df.isnull().sum()    
```
